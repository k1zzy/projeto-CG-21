### Um tutorial gráfico por André Falcão


Nesta aula focaremos 3 pontos do pipeline gráfico:
1. Grafo de cena usando a `glPushMAtrix()` e `glPopMatrix()` directamente no código (hard coded)
2. Grafo de cena usando uma estrutura flexível (Orientada a Objectos) que permite facil gestão da cena
