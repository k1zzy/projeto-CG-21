### Exercicios 1.

A primeira série de exercícios refere-se a à realização de animações simples com sprites usando mapeamento dinâmico de texturas. A cada frame há uma actualização das coordenadas UV da textura, permitindo ver uma nova imagem ligeiramente diferente da anterior. Sprites têm geralmente uma cor transparente que permite ver o que está no fundo e para que isso aconteça tem,os que ter activado um processamento de *alpha blending*

Vamos usar esta sequência de imagens retiradas  [daqui](https://blaiprat.github.io/jquery.animateSprite/) baseadas no cartoon [Scott Pilgrim](https://en.wikipedia.org/wiki/Scott_Pilgrim) 

![Diagram](scottpilgrim_1row.png)

1. Abra a imagem com um editor de imagens como o paint e verifique a cor transparente do *PNG* O Formato PNG suborta RGBA completo com um canal alpha completo
   1. Identifique as dimensões de cada frame (altura e largura)
2. Corra o código `1-simple-sprite.py` e examine-o. Carregando na barra de espaços o personagem dá um pequeno salto
   1. Veja se as medidas que determinou correspondem ao que o programa assume na construção de cada frame do sprite
   2. Há muitas secções que já lhe serão familiares mas Identifique onde a sprite é desenhada
   3. Veja como é processado o UV-mapping dinâmico de acordo com a frame
   4. Identifique as linas de código e as alterções que fazem com que a imagem ande para trás
   5. Veja com cuidado como é processado o salto (linhas 209-210). Experimente modificar o valor da gravidade
3. experimente rodar a câmara no eixo dos yy para ficar a olhar para a origem e observe como fica a sprite


### Exercicios 2.

Este segundo conjunto de exercícios é uma generalização do anterior, fazendo um simplificadíssimo *sprite engine* Criou-se uma extrutura orientada por objectos em que todos os objectos da cena terão métodos de `update()` e `draw()`, entre os quais, sprites. A cena básica tem duas sprites, baseadas na mesma imagem/textura, mas uma aparece mais longínqua e por isso menor e com outra velocidade
1. Corra o programa e observe o que se passa. Verifique os aretefactos presentes nesta versão e tente explicar o que se passa
2. Insira nas linhas 201-202 as seguintes instruções e corra de novo o programa, anotando o que ficou diferente
```
    glEnable(GL_ALPHA_TEST)
    glAlphaFunc(GL_GREATER, 0.1)   
``` 
    1. O que é que as linhas acimna fazem e porque é que os artefactos desaparecem?
    2. Haveria outra forma de resolver o problema?
3. Identifique a estrutura hierarquica da cena e como esta é definida no código
4. Crie novos "scotts" a correrem em zonas diferentes do ecrã a velocidades diferentes. Alguns muito próximos da câmara
5. Faça com que os "scotts" todos da imagem saltem simultaneamente quando pressiona a barra de espaços
    1. e com que saltem com velocidades e alturas diferentes

### Exercicios 3.

O 3º conjunto de exercícios é essencialmente descritivo. O propósito é usar e entender o processo de leitura de uma imagem OBJ [formato open soure do programa *Wavefront*] muito simpes e popular para distribuição de modelos. Este formato não contem o grafo de cena. Este terá que ser construído posteriormente com a informação aqui disponibilzada. O modelo que foi usado foi obtido [daqui](https://free3d.com/3d-model/grimlin-4802.html).

Uma questão que foi procurada realizar foi o uso de VBOs/VAOs, dado que este modelos têm frequenetemente dezenas ou centenas de milhares de polígonos. 

O programa `3-obj-reader.py` consegue fazer zoom usando as setas Up/Dn

1. Abra o ficheiro obj e mtl
   1. Identifique os elementos estruturais: vértices, normais, faces e texturas
   2. Verifique que elementos são apresentados no ficheiro .mtl
   3. Onde são definidas as texturas?
   4. Identifique as várias componentes do modelo
   5.  No ficheiro .mtl está a seguinte informação:
```
Ka 0 0 0 
Ns 1 
Ks 0.0784314 0.0784314 0.0784314 
Kd 0.0784314 0.0784314 0.0784314
```
       * o que quer dizer cada linha/componente deste fragmento
       * Descreva por palavras suas o que se está a descrever e qualifique-o
       * Porque são necessários vários blocos semelhantes?
2. Execute o programa e identifique:
   1. Que parte do programa armazena a geometria? Indique as linhas
   2. Que parte do programa armazena os materiais? Indique as linhas
   3. Existe alguma fonte de iluminação? Consegue criar mais uma?
3. Como conseguiria que o corpo do *grimlin* (e apenas o corpo, sem contar com braços ou cabeça), ficasse altamente refractora como um metal?
4. Porque razão estamos a usar VBOs? A GPU não é igualmente rápida a representar os mesmos triângulos com `glBegin()` e `glEnd()`?
5. Comente: Como faria para conseguir alguma animação neste personagem? Em particular braços e pernas. - Não é necessário fazê-lo, mas é importante perceber se, com este formato, isso é uma possibilidade
