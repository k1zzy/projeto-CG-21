### Instalação das pibliotecas que usaremos:

começaremos com a instalação do PyOpenGL e do glfw

`pip install PyOpenGL PyOpenGL_accelerate glfw`

*NOTA:* instalaremos a biblioteca `glfw` que substitui o GLUT, mas não usaremos ainda nesta aula. Ambas as bibliotecas servem para criar janelas, gerir contexto OpenGL e capturar input. O GLUT está mais indicado para o Pipline fixo do OpenGL e o GLFW é mais moderno e flexível. O GLUT será mais simples numa primeira fase, por isso usaremos esta biblioteca nesta semana. Nas próximas aulas usaremos o GLFW, com alterações mínimas no código

### Exercícios

Nos exercícios seguintes, a ideia é, **abrir um documento qualquer, e guardar directamente os screenshots do resultado obtido**, para serem mostrados ao docente na aula 

1. Veja com atenção a estrutura do código. O OpenGL funciona como uma máquina de estados e, através do GLUT as funções que fazem o processamento da cena são callbacks
2. Consegue identificar onde estão os parâmetros da câmara? E os parâmetros que definem a perspectiva?
3. Consegue identificar as transformações geométricas antes de desenhar o cubo? O que querem dizer?
4. Execute as seguintes transformações - para cada uma guarde o screenshot - **De notar que em OpenGL a última transformação é executada primeiro!**:
    1. Desloque o cubo para cima 3 unidades
    2. Desloque o cubo para cima 1 unidade e rode-o +25 graus no eixo dos x
    3. Desloque o cubo para baixo -1 unidade e rode-o -128 graus no eixo dos ys (atenção à ordem)
    4. escale o cubo para mais 50% e rode-o 30º no eixo dos zz
    5. faça um escalamento de 20% no eixo dos y e 50% no eixo dos zz, coloque-o em (1,-1,0) e rode-o 30 graus no eixo dos zz
    6. Ao objecto proveniente do escalamento anterior, desloque-o para (-1,-1, 0.5) e rode-o sobre o eixo (1,-1,0.5) 90 graus
5. desloque a câmara para a posição (-1.5, -2.0, 2.0) e repita as operações 4.A, 4.B e 4.E
6. Com a câmara nessa posição, desloque o ponto de observação para (1,1,1)
    1. Faça variar a inclinação da verticalidade da câmara 
7. veja o que acontece quando modificamos o fovy da perspectiva. experimente fovy de 30, 90 e 120 graus (o 60 é o valor por omissão) para os exercícios 4.A, 4.B e 4.E
8. Adicione uma pirâmide à cena. Esta deverá ser criada com 1 `GL_QUADS` para a base e 4 `GL_TRIANGLES` para os vértices. A pirâmide será criada com o centro em (0,0,0), mas para ser desenhada deverá estar com uma translação para (2,0, -1)
   1. a Pirâmide pode ser criada com uma função `faz_piramide(lado=1, altura=2)` onde se definem os lados da pirâmide e a altura do ápice
9. Com a pirâmide colocada na posição, experimente as transformações 4.A e 4.B aplicadas **simultaneamente ao cubo e à pirâmide**
10. Com a pirâmide colocada na posição, experimente as transformações 4.A e 4.B aplicadas apenas **ao cubo** enquanto que a pirâmide se mantém no mesmo sítio (esta pode ser mais complicada - como sugestão veja a função [glLoadIdentity()](https://docs.gl/gl2/glLoadIdentity))
11. Da **TP04** veja o **conjunto de exercícios 3** e:
     1. procure reproduzir em OpenGL o código que representa os 3 objectos (Exercício 3)
     2. Resolva o exercício 3.B em OpenGL
     3. Compare e discuta o procedimento feito na aula passada, nesta aula, e os resultados obtidos
